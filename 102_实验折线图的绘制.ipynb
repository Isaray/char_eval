```python
import pandas as pd 
import numpy as np
import json 
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('seaborn')
import pandas as pd 
def draw_line_chart(x,y_list,legned_list,xlabel,ylabel,title,save_dir="pic/"):
    plt.figure(figsize=(10, 6), dpi=100)
    for i in y_list:
        plt.plot(x, i)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(legned_list)
    plt.title(title)
    plt.savefig(save_dir+title+".png")
    plt.show()
```
